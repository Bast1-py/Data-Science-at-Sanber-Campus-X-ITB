```python
def check_values(df):
    data = []
    for col in df.columns:
        data.append([col, \
                  df[col].dtype, \
                  df[col].isna().sum(), \
                  round(100*(df[col].isna().sum()/len(df)), 2), \
                  df[col].nunique(), \
                  df[col].unique()
                  ])

    return pd.DataFrame(columns=['kolom', 'dataType', 'null', 'nullPercentage', 'unique', 'values'], data=data)
```

```python
check_values(df)
```

# Hands-On

```python
'''
Mengubah date menjadi tipedata datetime
'''

df['date'] = pd.to_datetime(df['date'], dayfirst=True, errors='coerce')
```

```python
check_values(df)
```

```python
'''
Membuat kolom baru tempdiff yang berisi perbedaan rata-rata hari ini dan sebelumnya
'''

df['tempdiff'] = df['meantempm'].diff() # Feature Engineering
```

```python
df['tempdiff']
```

```python
'''
Melakukan grouping quarterly dan mengetahui rata-rata temp terendah, tertinggi, dan rata-ratanya.
'''

result = df.groupby(pd.Grouper(key='date', freq='Q')).agg({
    'meantempm': ['min', 'max', 'mean']
}).reset_index()
```

```python
result
```

```python
df.index = df.date
```

```python
'''
Visualisasikan data dari waktu ke waktu secara seasonal

Tipe visualisasi:
Alasan:
Interpretasi:

'''

from statsmodels.graphics.tsaplots import quarter_plot

resampled = df['meantempm'].resample('Q').mean()
quarter_plot(resampled)
```

```python
df['Year'] = df['date'].dt.year
df['Month'] = df['date'].dt.month
```

```python
df.columns
```

```python
df
```

```python
temp = df.groupby(['Year', 'Month'])['meantempm'].mean()
temp
```

```python
meantemp_marix = df.groupby(['Year', 'Month'])['meantempm'].mean().unstack()
meantemp_marix
```

```python
'''
Visualisasikan data dari waktu ke waktu menggunakan heatmap

Tipe visualisasi:
Alasan:
Interpretasi:

'''

plt.figure(figsize=(12, 8))
sns.heatmap(meantemp_marix, cmap="YlGnBu", annot=True, fmt=".2f", linewidths=.5)

plt.xlabel('Month')
plt.ylabel('Year')
plt.title('Mean Temperature Heatmap')

plt.show()
```
