# Python

```python
import datetime
```

```python
date = datetime.date(year=2013, month=6, day=7)
time = datetime.time(hour=12, minute=30, second=19, microsecond=463198)
dt = datetime.datetime(year=2013, month=6, day=7, hour=12, minute=30, second=19, microsecond=463198)
```

```python
print("date is ", date)
```

```python
print("time is", time)
```

```python
print("datetime is", dt)
```

```python
'''
timedelta
'''

td = datetime.timedelta(
    weeks=2,
    days=5,
    hours=10,
    minutes=20,
    seconds=6.73,
    milliseconds=99,
    microseconds=8
    )

print(td)
```

```python
'''
Tambahkan timedelta ke date dan objek datetime
'''

print("date is ", date)
print('new date is', date + td)
```

```python
print("datetime is", dt)
print('new datetime is', dt + td)
```

# Shifting for Lagged Data

```python
microsoft.assign(
    prior_close = lambda x: x.Close.shift(),
    after_hours_change_in_price=lambda x: x.Open - x.prior_close,
    abs_change= lambda x: x.after_hours_change_in_price.abs()).nlargest(5, 'abs_change')
```

## Time Merging

```python
quotes = pd.DataFrame(
    {
        "time": [
            pd.Timestamp("2016-05-25 13:30:00.023"),
            pd.Timestamp("2016-05-25 13:30:00.023"),
            pd.Timestamp("2016-05-25 13:30:00.030"),
            pd.Timestamp("2016-05-25 13:30:00.041"),
            pd.Timestamp("2016-05-25 13:30:00.048"),
            pd.Timestamp("2016-05-25 13:30:00.049"),
            pd.Timestamp("2016-05-25 13:30:00.072"),
            pd.Timestamp("2016-05-25 13:30:00.075")
        ],
        "ticker": [
               "GOOG",
               "MSFT",
               "MSFT",
               "MSFT",
               "GOOG",
               "AAPL",
               "GOOG",
               "MSFT"
           ],
           "bid": [720.50, 51.95, 51.97, 51.99, 720.50, 97.99, 720.50, 52.01],
           "ask": [720.93, 51.96, 51.98, 52.00, 720.93, 98.01, 720.88, 52.03]
    }
)

quotes
```

```python
trades = pd.DataFrame(
       {
           "time": [
               pd.Timestamp("2016-05-25 13:30:00.023"),
               pd.Timestamp("2016-05-25 13:30:00.040"),
               pd.Timestamp("2016-05-25 13:30:00.042"),
               pd.Timestamp("2016-05-25 13:30:00.048"),
               pd.Timestamp("2016-05-25 13:30:00.048"),
               pd.Timestamp("2016-05-25 13:30:00.050")
           ],
           "ticker": ["MSFT", "MSFT", "MSFT", "GOOG", "GOOG", "AAPL"],
           "price": [51.95, 51.95, 51.95, 720.77, 720.92, 98.0],
           "quantity": [75, 155, 160, 100, 100, 100]
       }
   )
trades
```

```python
quotes[quotes['ticker'] == 'AAPL']
```

```python
pd.merge_asof(trades, quotes, on="time", by="ticker")
```

```python
pd.merge_asof(
    trades,
    quotes,
    on="time",
    by="ticker",
    tolerance=pd.Timedelta("10ms"),
    allow_exact_matches=False
)
```

```python
pd.merge_ordered(trades.reset_index(), quotes.reset_index()).set_index('time')
```
