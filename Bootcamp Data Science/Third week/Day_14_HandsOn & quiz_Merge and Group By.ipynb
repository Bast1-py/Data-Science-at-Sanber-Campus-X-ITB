## Read files

```python
import pandas as pd

inparts_df = pd.read_csv('inventory_parts.csv')
colors_df = pd.read_csv('colors.csv')
parts_df = pd.read_csv('parts.csv')
partscat_df = pd.read_csv('part_categories.csv')

partscat_df
```

```python
inparts_df.info()
colors_df.info()
parts_df.info()
partscat_df.info()
```

# Rename columns

```python
colors_df.rename(columns={"id": "color_id"}, inplace=True)
partscat_df.rename(columns={"id": "part_cat_id"}, inplace=True)

colors_df.info()
partscat_df.info()
```

# Merge data

```python
df = pd.merge(inparts_df,colors_df,on='color_id',how='inner')
df = pd.merge(df,parts_df,on='part_num',how='inner')
df = pd.merge(df,partscat_df,on='part_cat_id',how='inner')

df
```

# Rename again for better readibility

```python
df.rename(columns={"name_x": "color_name"}, inplace=True)
df.rename(columns={"name_y": "part_name"}, inplace=True)
df.rename(columns={"name": "part_cat_name"}, inplace=True)

df
```

# Group by

```python
# remove unnecessary columns for better readibility
df = df.drop(columns=df.columns[[0, 2, 9]])
```

```python
df.groupby(['part_cat_name', 'color_name']).sum(numeric_only=True)
```
