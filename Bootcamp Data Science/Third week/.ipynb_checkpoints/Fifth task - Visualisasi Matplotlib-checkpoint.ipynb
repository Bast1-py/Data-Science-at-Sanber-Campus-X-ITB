Data Set: Retail Sales Dataset

1. Transaction ID: A unique identifier for each transaction, allowing tracking and reference.
2. Date: The date when the transaction occurred, providing insights into sales trends over time.
3. Customer ID: A unique identifier for each customer, enabling customer-centric analysis.
4. Gender: The gender of the customer (Male/Female), offering insights into gender-based purchasing patterns.
5. Age: The age of the customer, facilitating segmentation and exploration of age-related influences.
6. Product Category: The category of the purchased product (e.g., Electronics, Clothing, Beauty), helping understand product preferences.
7. Quantity: The number of units of the product purchased, contributing to insights on purchase volumes.
8. Price per Unit: The price of one unit of the product, aiding in calculations related to total spending.
9. Total Amount: The total monetary value of the transaction, showcasing the financial impact of each purchase.

Source: https://www.kaggle.com/datasets/mohammadtalib786/retail-sales-dataset

```python
'''
Import modul yang dibutuhkan
'''
# Tuliskan Kode di sini
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
'''
Read file CSV
'''

# Tuliskan Kode di sini
dataset = pd.read_csv("retail_sales_dataset.csv")
dataset
```

```python
dataset.head(11)
```

```python
dataset.describe()
```

```python
dataset.nunique()
```

```python
dataset.Date = pd.to_datetime(dataset.Date,format = '%Y-%m-%d')
dataset['year'] = dataset.Date.dt.year
dataset['month'] = dataset.Date.dt.month
dataset['week'] = dataset.Date.dt.week
dataset['day'] = dataset.Date.dt.day
```

```python
dataset.head()
```

```python
dataset_year = dataset.groupby('year').sum().reset_index()
dataset_year.head()
```

```python
dataset_monthly = dataset.groupby('month').sum().reset_index()
dataset_monthly.head()
```

```python
dataset_weekly = dataset.groupby('week').sum().reset_index()
dataset_weekly.head()
```

```python
dataset_day = dataset.groupby('day').sum().reset_index()
dataset_day.head()
```

```python
dataset.describe()
```

### Soal 1

Visualisasikan distribusi setiap fitur/variabel

```python
# Tuliskan Kode di sini
plt.figure(figsize=(7, 16))
plt.style.use('ggplot')
ax0 = plt.subplot(511)
ax0 = plt.plot(dataset_year['year'], dataset_year['Total Amount'], label='Total Amount', marker='o')
plt.title('Line Chart: Retail Sales')
plt.xlabel('Year')
plt.ylabel('Total Amount')
plt.legend()
ax1 = plt.subplot(512)
ax1 = plt.plot(dataset_monthly['month'], dataset_monthly['Total Amount'], label='Total Amount', color='orange', marker='o')
plt.xlabel('Month')
plt.ylabel('Total Amount')
plt.legend()
ax2 = plt.subplot(513)
ax2 = plt.plot(dataset_weekly['week'], dataset_weekly['Total Amount'], label='Total Amount', color='blue', marker='o')
plt.xlabel('Week')
plt.ylabel('Total Amount')
plt.legend()
ax3 = plt.subplot(514)
ax3 = plt.plot(dataset_day['day'], dataset_day['Total Amount'], label='Total Amount', color='green', marker='o')
plt.xlabel('Day')
plt.ylabel('Total Amount')
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(7, 18))
ax0 = plt.subplot(511)
ax0 = plt.hist(dataset['Age'], histtype="stepfilled", bins=20, color='red', edgecolor='black')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Customer Age')

ax1 = plt.subplot(512)
ax1 = plt.hist(dataset['Quantity'],  bins=20, color='red', edgecolor='black')
plt.xlabel('Quantity')
plt.ylabel('Frequency')

ax2 = plt.subplot(513)
ax2 = plt.hist(dataset['Total Amount'], histtype="stepfilled", bins=20, color='red', edgecolor='black')
plt.xlabel('Total Amount')
plt.ylabel('Frequency')

ax3 = plt.subplot(514)
ax3 = plt.hist(dataset['Product Category'], bins=20, color='red', edgecolor='black')
plt.xlabel('Product Category')
plt.ylabel('Frequency')

plt.show()
```

### Soal 5

Bagaimana tren penjualan dari waktu ke waktu? Apakah terdapat peningkatan di suatu tanggal tertentu?


Dari visualisasi line chart diatas, saya membuat visualisasi per bulan, per minggu, dan per hari.  
untuk satu tahun yang mana jumlah transaksi dan quantity terbanyak terjadi pada bulan ke-lima dan mengalamai penurunan cukup banyak di bulan ke-sembilan. 
Dan untuk satu bulan tersebut paling banyak terjadi pada range tanggal 20 - 25.


```python
dataset['Date'] = pd.to_datetime(dataset['Date'])

daily_sales = dataset.groupby('Date')['Total Amount'].sum().reset_index()

plt.figure(figsize=(12, 6))
plt.plot(daily_sales['Date'], daily_sales['Total Amount'], marker='o', linestyle='-')
plt.title('Tren Penjualan Harian')
plt.xlabel('Tanggal')
plt.ylabel('Total Penjualan')
plt.grid(True)
plt.show()
```

```python
dataset['Month'] = dataset['Date'].dt.to_period('M')

monthly_sales = dataset.groupby('Month')['Total Amount'].sum()

plt.figure(figsize=(12, 6))
plt.plot(monthly_sales.index.astype(str), monthly_sales.values, marker='o', linestyle='-')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
