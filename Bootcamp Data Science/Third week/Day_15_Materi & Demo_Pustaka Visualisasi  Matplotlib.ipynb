# How to Capture a Trend

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generating sample data
np.random.seed(42)

# Simulating time periods
time_periods = pd.date_range(start='2022-01-01', end='2022-12-31', freq='M')

# Simulating revenue data
revenue_data = np.cumsum(np.random.randint(50000, 100000, size=len(time_periods)))

# Simulating Apple and Amazon stock prices
apple_stock = np.cumsum(np.random.normal(50, 20, size=len(time_periods)))
amazon_stock = np.cumsum(np.random.normal(70, 30, size=len(time_periods)))

# Simulating total sales data for area chart
total_sales_data = np.cumsum(np.random.randint(1000, 5000, size=len(time_periods)))

# Simulating active users by segment for stacked area chart
segment_a = np.random.randint(100, 500, size=len(time_periods))
segment_b = np.random.randint(50, 300, size=len(time_periods))
segment_c = np.random.randint(20, 100, size=len(time_periods))

# Simulating electricity consumption over time for spline chart
electricity_consumption = np.cumsum(np.random.randint(500, 1000, size=len(time_periods)))

# Creating a DataFrame
df = pd.DataFrame({
    'Time': time_periods,
    'Revenue': revenue_data,
    'Apple_Stock': apple_stock,
    'Amazon_Stock': amazon_stock,
    'Total_Sales': total_sales_data,
    'Segment_A': segment_a,
    'Segment_B': segment_b,
    'Segment_C': segment_c,
    'Electricity_Consumption': electricity_consumption
})
```

```python
# Line Chart - Revenue over Time
plt.figure(figsize=(5, 3))
plt.plot(df['Time'], df['Revenue'], label='Revenue', marker='o')
plt.title('Line Chart: Revenue over Time')
plt.xlabel('Time')
plt.ylabel('Revenue ($)')
plt.legend()
plt.show()
```

```python
# Multi-Line Chart - Apple vs Amazon Stocks over Time
plt.figure(figsize=(5, 3))
plt.plot(df['Time'], df['Apple_Stock'], label='Apple Stock', marker='o')
plt.plot(df['Time'], df['Amazon_Stock'], label='Amazon Stock', marker='o')
plt.title('Multi-Line Chart: Apple vs Amazon Stocks over Time')
plt.xlabel('Time')
plt.ylabel('Stock Price')
plt.legend()
plt.show()
```

```python
# Area Chart - Total Sales over Time
plt.figure(figsize=(5, 3))
plt.fill_between(df['Time'], df['Total_Sales'], color='skyblue', alpha=0.4)
plt.title('Area Chart: Total Sales over Time')
plt.xlabel('Time')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Stacked Area Chart - Active Users over Time by Segment
plt.figure(figsize=(5, 3))
plt.stackplot(df['Time'], df['Segment_A'], df['Segment_B'], df['Segment_C'], labels=['Segment A', 'Segment B', 'Segment C'], alpha=0.7)
plt.title('Stacked Area Chart: Active Users over Time by Segment')
plt.xlabel('Time')
plt.ylabel('Active Users')
plt.legend()
plt.show()
```

# How to Visualize Relationships

```python
# Bar Chart - Volume of Google Searches by Region

data = {'Region': ['North', 'South', 'East', 'West'],
        'SearchVolume': [500, 300, 700, 400]}
google_search_data = pd.DataFrame(data)

plt.figure(figsize=(5, 3))
plt.bar(google_search_data['Region'], google_search_data['SearchVolume'])
plt.title('Volume of Google Searches by Region')
plt.xlabel('Region')
plt.ylabel('Search Volume')
plt.show()
```

```python
# Column Chart - Brand Market Share

data = {'Brand': ['Brand A', 'Brand B', 'Brand C'],
        'MarketShare': [30, 40, 30]}
market_share_data = pd.DataFrame(data)

plt.figure(figsize=(5, 3))
plt.bar(market_share_data['Brand'], market_share_data['MarketShare'], color=['blue', 'green', 'red'])
plt.title('Brand Market Share')
plt.xlabel('Brand')
plt.ylabel('Market Share (%)')
plt.show()
```

```python
# Scatter Plot - Relationship between Salary and Years Spent at Company

data = {'YearsAtCompany': [2, 5, 8, 3, 7],
        'Salary': [50000, 60000, 75000, 55000, 70000]}
employee_data = pd.DataFrame(data)

plt.figure(figsize=(5, 3))
plt.scatter(employee_data['YearsAtCompany'], employee_data['Salary'])
plt.title('Relationship between Salary and Years Spent at Company')
plt.xlabel('Years at Company')
plt.ylabel('Salary')
plt.show()
```

```python
# Connected Scatter Plot - Cryptocurrency Price Index

data = {'Date': ['2024-01-01', '2024-01-02', '2024-01-03'],
        'BitcoinPrice': [50000, 52000, 49000],
        'EthereumPrice': [1500, 1550, 1480]}
crypto_data = pd.DataFrame(data)

plt.figure(figsize=(5, 3))
plt.plot(crypto_data['Date'], crypto_data['BitcoinPrice'], marker='o', label='Bitcoin')
plt.plot(crypto_data['Date'], crypto_data['EthereumPrice'], marker='o', label='Ethereum')
plt.title('Cryptocurrency Price Index')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()
```

```python
# Bubble Chart - Relationship between Life Expectancy, GDP per Capita, Population Size

data = {'Country': ['USA', 'China', 'India', 'Brazil'],
        'LifeExpectancy': [78, 76, 69, 75],
        'GDPCapita': [60000, 10000, 5000, 12000],
        'Population': [330000000, 1400000000, 1400000000, 210000000]}
world_data = pd.DataFrame(data)

plt.figure(figsize=(5, 3))
plt.scatter(world_data['GDPCapita'], world_data['LifeExpectancy'], s=world_data['Population']/1000000, alpha=0.5)
plt.title('Relationship between Life Expectancy, GDP per Capita, and Population')
plt.xlabel('GDP per Capita')
plt.ylabel('Life Expectancy')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data contoh
bulan = ['Jan', 'Feb', 'Mar', 'Apr', 'Mei']
penjualan_unit = [100, 120, 90, 110, 95]
pendapatan = [5000, 6000, 4500, 5500, 4750]

# Membuat objek figure dan axis
fig, ax1 = plt.subplots()

# Plotting penjualan unit menggunakan sumbu y pertama
ax1.set_xlabel('Bulan')
ax1.set_ylabel('Penjualan Unit', color='tab:blue')
ax1.bar(bulan, penjualan_unit, color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Membuat sumbu y kedua untuk pendapatan
ax2 = ax1.twinx()
ax2.set_ylabel('Pendapatan', color='tab:red')
ax2.plot(bulan, pendapatan, color='tab:red', marker='o')
ax2.tick_params(axis='y', labelcolor='tab:red')

# Menampilkan grafik
plt.title('Penjualan Unit dan Pendapatan per Bulan')
plt.show()
```

```python
# Word Cloud Chart - Top 10 Words by Customers

from wordcloud import WordCloud

data = {'CustomerID': [1, 2, 3, 4, 5],
        'ReviewText': ['Excellent service', 'Product is amazing', 'Could be better', 'Fast delivery', 'Highly recommended']}
customer_reviews = pd.DataFrame(data)

# Combine all reviews into a single string
text = ' '.join(customer_reviews['ReviewText'])

# Generate Word Cloud
wordcloud = WordCloud(width=800, height=400, max_words=10, background_color='white').generate(text)

# Display the Word Cloud
plt.figure(figsize=(5, 3))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Top 10 Words by Customers')
plt.show()
```

# How to Visualize Part-to-whole Charts

```python
# Pie Chart - Expenses by Category

# Sample Data
data = {'Category': ['Food', 'Transportation', 'Entertainment'],
        'Expenses': [500000, 200000, 700000]}
exp_data = pd.DataFrame(data)

plt.pie(exp_data['Expenses'], labels=exp_data['Category'], autopct='%1.1f%%', startangle=90)
plt.title('Expenses by Category')
plt.show()
```

```python
# Donut Pie Chart - Android OS Market Share

data = {'AndroidVersion': ['Pie', 'Oreo', 'Nougat', 'Marshmallow'],
        'MarketShare': [40, 30, 15, 15]}
android_data = pd.DataFrame(data)

plt.pie(android_data['MarketShare'], labels=android_data['AndroidVersion'], autopct='%1.1f%%', startangle=90, wedgeprops=dict(width=0.4))
plt.title('Android OS Market Share')
plt.show()
```

```python
# Heat Map - Average Monthly Temperatures Across the Year

data = {'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
        '2023': [10, 15, 20, 25, 30, 35],
        '2024': [12, 17, 22, 27, 32, 37]}
temperature_data = pd.DataFrame(data)

plt.imshow(temperature_data[['2023', '2024']], cmap='viridis', aspect='auto', interpolation='nearest')
plt.colorbar(label='Average Temperature (C)')
plt.yticks(range(len(temperature_data['Month'])), temperature_data['Month'])
plt.xticks(range(len(temperature_data.columns)-1), temperature_data.columns[1:])
plt.title('Average Monthly Temperatures Across the Year')
plt.show()
```

# How to Visualize Distribution

```python
# Set random seed for reproducibility
np.random.seed(42)

# Generate random data for normal distribution
data_normal = np.random.normal(size=1000)

# Generate random data for bimodal distribution
data_bimodal = np.concatenate([np.random.normal(loc=-2, size=500), np.random.normal(loc=2, size=500)])

# Generate random data for skewed distribution
data_skewed = np.random.exponential(size=1000)
```

```python
df_normal = pd.DataFrame({'Values': data_normal, 'Distribution': 'Normal'})
df_bimodal = pd.DataFrame({'Values': data_bimodal, 'Distribution': 'Bimodal'})
df_skewed = pd.DataFrame({'Values': data_skewed, 'Distribution': 'Skewed'})

df = pd.concat([df_normal, df_bimodal, df_skewed])
```

```python
# 1. Histogram
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.hist(df[df['Distribution'] == 'Normal']['Values'], bins=30, color='skyblue', edgecolor='black')
plt.title('Normal Distribution')

plt.subplot(1, 3, 2)
plt.hist(df[df['Distribution'] == 'Bimodal']['Values'], bins=30, color='lightcoral', edgecolor='black')
plt.title('Bimodal Distribution')

plt.subplot(1, 3, 3)
plt.hist(df[df['Distribution'] == 'Skewed']['Values'], bins=30, color='lightgreen', edgecolor='black')
plt.title('Skewed Distribution')

plt.tight_layout()
plt.show()
```

```python
# 2. Box plot
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.boxplot(df[df['Distribution'] == 'Normal']['Values'], vert=False)
plt.title('Normal Distribution')

plt.subplot(1, 3, 2)
plt.boxplot(df[df['Distribution'] == 'Bimodal']['Values'], vert=False)
plt.title('Bimodal Distribution')

plt.subplot(1, 3, 3)
plt.boxplot(df[df['Distribution'] == 'Skewed']['Values'], vert=False)
plt.title('Skewed Distribution')

plt.tight_layout()
plt.show()
```

```python
# 3. Violin plot
plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.violinplot(df[df['Distribution'] == 'Normal']['Values'], showextrema=False, vert=False)
plt.title('Normal Distribution')

plt.subplot(1, 3, 2)
plt.violinplot(df[df['Distribution'] == 'Bimodal']['Values'], showextrema=False, vert=False)
plt.title('Bimodal Distribution')

plt.subplot(1, 3, 3)
plt.violinplot(df[df['Distribution'] == 'Skewed']['Values'], showextrema=False, vert=False)
plt.title('Skewed Distribution')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)
data = np.random.gamma(shape=2, scale=2, size=1000)
df = pd.DataFrame({'Values': data})
df['Values'] = np.floor(df['Values']).astype(int)
data = df['Values'].values
data = df['Values'].dropna().values
```

```python
import statistics

mean_val = np.mean(data)
mode_val = statistics.multimode(data)
median_val = np.median(data)
```

```python
mean_val = df['Values'].mean()
mode_val = df['Values'].mode().iloc[0]
median_val = df['Values'].median()
```

```python
# Create a figure with two subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Plot KDE on the first subplot
sns.kdeplot(data, ax=axes[0], color='green', fill=True)
axes[0].axvline(mean_val, color='red', linestyle='dashed', linewidth=2, label='Mean')
axes[0].axvline(mode_val, color='blue', linestyle='dashed', linewidth=2, label='Mode')
axes[0].axvline(median_val, color='orange', linestyle='dashed', linewidth=2, label='Median')
axes[0].set_title('Kernel Density Estimate (KDE)')
axes[0].legend()

# Plot histogram on the second subplot
sns.histplot(data, ax=axes[1], kde=False, color='green', bins=30)
axes[1].axvline(mean_val, color='red', linestyle='dashed', linewidth=2, label='Mean')
axes[1].axvline(mode_val, color='blue', linestyle='dashed', linewidth=2, label='Mode')
axes[1].axvline(median_val, color='orange', linestyle='dashed', linewidth=2, label='Median')
axes[1].set_title('Histogram')
axes[1].legend()

# Adjust layout
plt.tight_layout()
plt.show()
```
