```python
import random

# Membuat data acak untuk contoh DataFrame
data = {
    'Nama': [f'Orang-{i}' for i in range(1, 21)],
    'Usia': [random.randint(18, 60) for _ in range(20)],
    'Kota': [random.choice(['Jakarta', 'Surabaya', 'Bandung', 'Medan', 'Semarang']) for _ in range(20)],
    'Pekerjaan': [random.choice(['Guru', 'Dokter', 'Insinyur', 'Pengacara', 'Desainer']) for _ in range(20)],
    'Pendapatan': [random.randint(2000000, 16000000) for _ in range(20)]
}

# Membuat DataFrame dari data
df = pd.DataFrame(data)

# Menampilkan DataFrame
print(df)
```
