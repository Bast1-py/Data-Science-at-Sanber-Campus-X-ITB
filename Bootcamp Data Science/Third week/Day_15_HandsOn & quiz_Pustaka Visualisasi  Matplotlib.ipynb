```python
def check_values(df):
    data = []
    for col in df.columns:
        data.append([col, \
                  df[col].dtype, \
                  df[col].isna().sum(), \
                  round(100*(df[col].isna().sum()/len(df)), 2), \
                  df[col].nunique(), \
                  df[col].unique()
                  ])

    return pd.DataFrame(columns=['kolom', 'dataType', 'null', 'nullPercentage', 'unique', 'values'], data=data)
```

```python
check_values(df)
```

```python
df.columns
```

```python
df.Date = pd.to_datetime(df.date,format = '%Y-%m-%d')
df['year'] = df.Date.dt.year
df['month'] = df.Date.dt.month
df['day'] = df.Date.dt.day

# Dokumentasi
# https://docs.python.org/3/library/datetime.html
```

```python
df['week'] = df.Date.dt.week
```

## 1. Tren Covid Bulanan

```python
df_monthly = df.groupby('month').sum()
df_monthly
```

```python
df_monthly = df.groupby('month').sum().reset_index()
df_monthly.head()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df_monthly['month'], df_monthly['cases'], label='Cases', marker='o')
plt.title('Line Chart: Cases over Time')
plt.xlabel('Month')
plt.ylabel('Cases')
plt.legend()
plt.show()
```

## 2. Tren Meninggal Bulanan

```python
df.tail()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df_monthly['month'], df_monthly['deaths'], label='deaths', marker='o')
plt.title('Line Chart: deaths over Time')
plt.xlabel('Month')
plt.ylabel('death')
plt.legend()
plt.show()
```

## 3. Tren Covid Mingguan

```python
df_weekly = df.groupby('week').sum().reset_index()
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df_weekly['week'], df_weekly['cases'], label='Cases', marker='o')
plt.title('Line Chart: Cases over Time')
plt.xlabel('Month')
plt.ylabel('Cases')
plt.legend()
plt.show()
```

## 4. Top 10 Country with The Most Cases

```python
top_10_counties = df.groupby('county')['cases'].sum().sort_values(ascending=False).head(10).reset_index()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(top_10_counties['county'], top_10_counties['cases'])
plt.title('Top 10 Countries with The Most Cases')
plt.xlabel('Country')
plt.ylabel('Cases')
plt.show()
```

## 5. Top 10 States with The Most Cases

```python
top_10_states = df.groupby('state')['cases'].sum().sort_values(ascending=False).head(10).reset_index()
top_10_states.head()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(top_10_states['state'], top_10_states['cases'])
plt.title('Top 10 states with The Most Cases')
plt.xlabel('state')
plt.ylabel('Cases')
plt.show()
```

## 10. Heatmap of Cases by Month and Day

```python
df_heatmap = df.pivot_table(index='month', columns='day', values='cases', aggfunc='sum')
plt.imshow(df_heatmap, cmap='YlOrRd', aspect='auto', interpolation='none')
plt.title('Heatmap of Cases by Month and Day')
plt.xlabel('Day')
plt.ylabel('Month')
plt.colorbar(label='Cases')
plt.show()
```
