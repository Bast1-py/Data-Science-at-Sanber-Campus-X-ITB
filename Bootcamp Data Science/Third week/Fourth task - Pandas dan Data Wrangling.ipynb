# 5.  Menentukan jenis data categorical atau numeric
Jelaskan data kategorikal (kualitatif) dan data numerik (kuantitatif)

Jawab disini:

data kategorikal (kualitatif) itu data yang disajikan dalam bentuk kata-kata (tulisan), gambar (audio) atau video yang memiliki makna. Data-data tersebut diperoleh dari wawancara, pengamatan, pemotretan, perekaman dan lain-lain. Pada intinya data kualitatif adalah data yang bukan merupakan bilangan angka sehinga tidak dianalisis dengan ilmu statistik (statistika).

sedangkan, data numerik (kuantitatif) itu data yang disajikan dalam bentuk bilangan angka, sehingga data kualitatif diolah secara statistik. Data kualitatif terbagi lagi menjadi dua jenis yaitu data kategorik dan data numerik. Data kategorikal ini terbagi menjadi nominal, ordinal, interval, dan rasio.
