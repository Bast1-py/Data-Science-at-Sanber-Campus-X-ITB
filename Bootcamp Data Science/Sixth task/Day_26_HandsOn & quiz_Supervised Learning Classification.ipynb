1. Buat model Naive Bayes, KNN, Decision Tree, dan SVM. Kemudian, bandingkan hasil evaluasi accuracy kelima model (termasuk logistic regression). Manakah yang lebih bagus?

```python
X = df.drop(["Outcome"], axis = 1) # Variabel independen
y = df["Outcome"]  # Variabel dependen / Label
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_scaled, y, test_size=0.2, random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import cross_val_score
import pandas as pd

classifiers = [
    ('LogReg', LogisticRegression()),
    ('SVM', SVC()),
    ('KNN', KNeighborsClassifier()),
    ('NB', GaussianNB()),
    ('DT', DecisionTreeClassifier())
]

result_cls = pd.DataFrame(columns=['Model', 'Accuracy'])

for name, model in classifiers:
    model.fit(X_train, y_train)
    accuracy = cross_val_score(model, X_train, y_train, cv=5, scoring='accuracy').mean()
    result_cls.loc[len(result_cls)] = [name, accuracy]

result_cls.sort_values(by='Accuracy', ascending=False)
```

```python
lg = LogisticRegression()
lg.fit(X_train,y_train)
y_pred = lg.predict(X_val)
acc = metrics.accuracy_score(y_val, y_pred)
acc
```

```python
from sklearn.model_selection import GridSearchCV, RandomSearchCV

params = {'C': [0.01, 0.1, 1, 10, 100],
          'penalty': ['l1', 'l2'],
          'max_iter': [100, 150],
          'solver': ['liblinear', 'saga']}

lr = LogisticRegression()
gs = GridSearchCV(lr, params, cv=5, scoring='accuracy')
gs.fit(X_train, y_train)
```

```python
best_params = gs.best_params_
print("Best Hyperparameters:", best_params)
```

```python
best_model = gs.best_estimator_

y_pred = best_model.predict(X_val)
acc = metrics.accuracy_score(y_val, y_pred)
acc
```

```python
for c in cluster_range:
    cntr, u, u0, d, jm, p, fpc = fuzz.cluster.cmeans(
        data.T, c, 2, error=0.005, maxiter=1000)

    cluster_membership = np.argmax(u, axis=0)

    silhouette_avg = silhouette_score(data, cluster_membership)
    silhouette_scores.append(silhouette_avg)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(8, 4))
plt.plot(cluster_range, silhouette_scores, marker='o', linestyle='-', color='b')
plt.title('Silhouette Score vs. Number of Clusters')
plt.xlabel('Number of Clusters')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.show()
```
