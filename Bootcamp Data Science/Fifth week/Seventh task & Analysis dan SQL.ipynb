3. Simpan tabel yang ada di dalam database ke dataframe pandas Python

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("harga.csv")
df
```

```python
df.Tanggal = pd.to_datetime(df.Tanggal,format = '%d-%m-%Y')
df['day'] = df.Tanggal.dt.day
df['month'] = df.Tanggal.dt.month
df['year'] = df.Tanggal.dt.year
df
```

4. Lakukan univariate analysis untuk variabel harga: Buat visualisasinya (bebas) dan jelaskan insight yang didapat dari visualisasi tersebut

```python
import seaborn as sns

plt.subplots(figsize = (16, 6))
am = sns.barplot(df.Harga, df.day)
plt.xlabel('Harga')
plt.ylabel('Day')
plt.title('Harga Tiket berdasarkan Jenis')
plt.show()
```

Dari visual yang saya buat, saya memilih tipe bar chart untuk merepresentasikan perbandingan Harga proporsi nilai dari suatu kategorikal dalam suatu variabel seperti Tanggal (Tahun, bulan, dan hari), Jenis, dan Kelas suatu database maskapai. Analisa univariate jenis ini merupakan analisa data feature tunggal yaitu proporsi perbandingan Harga.

```python
import seaborn as sns

plt.subplots(figsize = (18, 6))
am = sns.barplot(df.Jenis, df.Harga)
plt.xlabel('Jenis')
plt.ylabel('Harga')
plt.title('Harga Tiket berdasarkan Jenis')
plt.show()
```

5. Lakukan juga univariate analysis untuk variabel kelas: Buat visualisasinya (bebas) dan jelaskan insight yang didapat dari visualisasi tersebut

```python
import seaborn as sns


plt.subplots(figsize = (18, 6))
am = sns.barplot(df.Kelas, df.Harga)
plt.xlabel('Kelas')
plt.ylabel('Harga')
plt.title('Harga Tiket berdasarkan Kelas')
plt.show()
```

Pencitraan berdasarkan Variabel Kelas yang dilakukan untuk melihat proporsi harga tiket pesawat. Pencitraan yang saya buat dengan menggunakan bar chart untuk melihat perbandingan dari variabel Kelas. Bisa dilihat perbedaan menunjukkan bahwa Kelas bisnis memiliki proporsi nilai harga tiket lebih mahal ketimbang kelas Ekonomi.

6. Bandingkan harga dari tiap jenis maskapai dari waktu ke waktu. Visualisasikan.

```python
import seaborn as sns

plt.subplots(figsize = (18, 6))
am = sns.barplot(df.day, df.Harga)
plt.xlabel('Day')
plt.ylabel('Harga')
plt.title('Harga Tiket berdasarkan Tanggal (Hari)')
plt.show()
```

7. Group dataframe berdasarkan jenis harinya, week-end/week-day, kemudian diagregasikan menggunakan average untuk setiap maskapai

```python
import datetime
# mengubah Tanggal (date) menjadi tipedata datetime
df['Tanggal'] = pd.to_datetime(df['Tanggal'], dayfirst=True, errors='coerce')
df
```

```python
result = df.groupby(pd.Grouper(key='Tanggal')).agg({
    'Harga': ['min', 'max', 'mean']
}).reset_index()
result
```

```python
result = df.groupby(pd.Grouper(key='Jenis')).agg({
    'Harga': ['min', 'max', 'mean']
}).reset_index()
result
```

8. Lakukan bivariate analysis untuk variabel harga dan kelas: Buat visualisasinya (bebas) dan jelaskan insight yang didapat dari visualisasi tersebut

```python
with sns.axes_style(style='ticks'):
    g = sns.catplot(x="Kelas", y="Harga", data=df, kind="box")
    g.set_axis_labels("Kelas", "Harga")
    plt.show()
```

Untuk bivariate analysis dengan variabel harga dan kelas saya menggunakan distribusi box plot untuk menampilan hubungan index kelas yang kategorikal dengan harga tertinggi yang numerikal. Dilihat dari hasil plot bahwa kelas Bisnis memiliki harga cukup mahal jika dibandingkan kelas ekonomi
