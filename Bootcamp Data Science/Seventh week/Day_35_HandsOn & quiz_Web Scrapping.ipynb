# Scraping Wikipedia

```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
alamat = "https://en.wikipedia.org/wiki/List_of_brightest_stars"
html = urlopen(alamat)
data = BeautifulSoup(html, 'html.parser')
```

```python
alamat1 = "https://www.kompas.com/"
html1 = urlopen(alamat1)
data1 = BeautifulSoup(html1, 'html.parser')
```

```python
table = data1.findAll("table", {"class":"mostItem"})
table
```

```python
table = data.findAll("table", {"class":"wikitable"})[0]
table
```

```python
rows = table.find_all("tr")
rows
```

```python
hasil = []
for row in rows:
    info = []
    for cell in row.findAll(["td", "th"]):
        info.append(cell.get_text().strip())
    hasil.append(info)
```

# Scraping Google Play Store Reviews

```python
from google_play_scraper import app
import pandas as pd
import numpy as np
```

```python
from google_play_scraper import Sort, reviews_all


us_reviews = reviews_all(
    'com.busuu.android.en',
    sleep_milliseconds=0, # defaults to 0
    lang='en', # defaults to 'en'
    country='us', # defaults to 'us'
    sort=Sort.NEWEST, # defaults to Sort.MOST_RELEVANT
)
```

```python
df_busu = pd.DataFrame(np.array(us_reviews),columns=['review'])
df_busu = df_busu.join(pd.DataFrame(df_busu.pop('review').tolist()))
df_busu.head()
```
